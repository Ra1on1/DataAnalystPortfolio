```python
#Importieren von Bibliotheken
import ast
import pandas as pd
import seaborn as sns
from datasets import load_dataset
import matplotlib.pyplot as plt

#Daten laden
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

#Datenbereinigung
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
df['job_skills'] = df['job_skills'].apply(lambda x: ast.literal_eval(x) if pd.notna(x) else x)
```

```python
df_DA_GER = df[(df['job_title'] == 'Data Analyst') & (df['job_country'] == 'United States')].copy()
```

```python
df_DA_GER['job_posted_month_no'] = df_DA_GER['job_posted_date'].dt.month
```

```python
df_DA_US_explode = df_DA_GER.explode('job_skills')
```

```python
df_DA_GER_pivot =  df_DA_US_explode.pivot_table(index='job_posted_month_no', columns='job_skills', aggfunc='size', fill_value=0)

df_DA_GER_pivot.loc['Total'] = df_DA_GER_pivot.sum()

df_DA_GER_pivot = df_DA_GER_pivot[df_DA_GER_pivot.loc['Total'].sort_values(ascending=False).index]

df_DA_GER_pivot = df_DA_GER_pivot.drop('Total')

df_DA_GER_pivot
```

```python
DA_totals = df_DA_GER.groupby('job_posted_month_no').size()
```

```python
df_DA_GER_percent = df_DA_GER_pivot.div(DA_totals/100, axis=0)

df_DA_GER_percent
```

```python
# Index zurücksetzen, um sicherzustellen, dass 'job_posted_month_no' eine Spalte ist
df_DA_GER_percent = df_DA_GER_percent.reset_index()

# Monatsnummer in Monatsname umwandeln
df_DA_GER_percent['job_posted_month'] = df_DA_GER_percent['job_posted_month_no'].apply(
    lambda x: pd.to_datetime(str(x), format='%m').strftime('%b')
)

# Löscht 'job_posted_month_no' bevor der index gesetzt wird
df_DA_GER_percent = df_DA_GER_percent.drop(columns='job_posted_month_no')

# Setzt 'job_posted_month' als den neuen index
df_DA_GER_percent = df_DA_GER_percent.set_index('job_posted_month')

df_DA_GER_percent
```

```python
df_plot = df_DA_GER_percent.iloc[:, :5]

sns.lineplot(data=df_plot, dashes=False, palette='tab10')
sns.set_theme(style='ticks')
sns.despine()

plt.title('Top Qualifikationen für Datenanalysten in Deutschland')
plt.ylabel('Wahrscheinlichkeit in der Stellenausschreibung')
plt.xlabel('2023')
plt.legend().remove()

from matplotlib.ticker import PercentFormatter
ax = plt.gca()
ax.yaxis.set_major_formatter(PercentFormatter(decimals=0))


for i in range(5):
    plt.text(11.2, df_plot.iloc[-1, i], df_plot.columns[i])
```
